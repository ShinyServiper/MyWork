# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re

import nltk
nltk.download('reuters')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Expand the default stopwords list if necessary
# YOUR CODE HERE!
sw = set(stopwords.words('english')) 
sw_addons = {'char','chars','reuters','Reuters'}

```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text. The function performs the following tasks
    # Create a list of the words
    # Convert the words to lowercase
    # Remove the punctuation
    # Remove the stop words
    # Lemmatize Words into root words 
    """
    
    sw = set(stopwords.words('english'))                              #creating set of stopwords
    lemmatizer = WordNetLemmatizer()                                  #creating lemmatizer object
    regex = re.compile("[^a-zA-Z ]")                                  #defining regular expression
    re_clean = regex.sub('', text)                                    #removing nonalphanumeric chars
    words = word_tokenize(re_clean)                                   #turning text into words
    lem = [lemmatizer.lemmatize(word) for word in words]              #lemmatize words
    tokens = [word.lower() for word in lem if word.lower() not in sw.union(sw_addons)] #filtering out stopwords
    return tokens                                                     #returning processed text
```

```python
# Create a new tokens column for bitcoin
# YOUR CODE HERE!
btc_df['tokens']=[tokenizer(text) for text in btc_df['text']]
btc_df.head()
```

```python
# Create a new tokens column for ethereum
# YOUR CODE HERE!
eth_df['tokens']=[tokenizer(text) for text in eth_df['text']]
eth_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams

def gram_count(df, N=2):
    """constructs dictionary of total n-gram counts from provided DataFrame"""
    gram_dict={}                                          #initialized empty dictionary
    for tokens in df["tokens"]:
        bigram_counts = Counter(ngrams(tokens,N))   #gets ngrams
        bigram_counts=dict(bigram_counts)                 #converts bigram info into dictionary
        gram_dict.update(bigram_counts)                   #updates dictionary with new entries
    return gram_dict                                      #returning dictionary of ngrams
```

# Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
# Generate the Bitcoin word cloud
# YOUR CODE HERE!
def CloudCreate(df,my_title):
    words=[]                               #initial empty list of words
    for tokens in df:
        words=words+tokens                 #adding to the list
    big_string = ' '.join(words)           #make one big stringn separated by white spaces 
    wc = WordCloud().generate(big_string)  #constructs word cloud setup
    plt.axis('off')
    plt.title(my_title)
    plt.imshow(wc)                         #displays wordcloud
    
#Displaying the WordCloud
CloudCreate(btc_df["tokens"],"Bitcoin WordCloud")
```

```python
# Generate the Ethereum word cloud
# YOUR CODE HERE!
CloudCreate(eth_df["tokens"],"Ethereum WordCloud")
```
